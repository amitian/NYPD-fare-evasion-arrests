```python
median_incomes = pd.concat([manhattan_L, brooklyn_L], axis=0).reset_index(drop=True)
```

```python
stations = {
        '007700' : '8th Av', 
        '005400' : '6th Av',
        '005000' : 'Union Square',
        '004800' : '3rd Av',
        '004400' : '1st Av',
        '055300' : 'Bedford Av',
        '050300' : 'Lorimer St',
        '049700' : 'Graham Av',
        '048500' : 'Grand, Morgan',
        '049300' : 'Montrose Av',
        '044700' : 'Jefferson St',
        '044500' : 'DeKalb Av',
        '043900' : 'Myrtle, Halsey',
        '040900' : 'Wilson Av',
        '040500' : 'Bushwick Av',
        '036700' : 'Broadway Jct',
        '114400' : 'Atlantic Av',
        '090800' : 'Sutter Av',
        '113200' : 'Livonia Av',
        '092000' : 'New Lots Av',
        '109800' : 'East 105th St',
        '097400' : 'Canarsie'}
```

```python
def find_stations(row, mapper):
    for k, v in mapper.items():
        if k == row:
            return v
    else:
        return None

median_incomes['STATION'] = median_incomes.TRACT.apply(find_stations, mapper=stations)
```

```python
median_incomes.rename(columns= {'DP03_0086E' : 'MEDIAN_INCOME'}, inplace = True)
```

```python
median_incomes['MEDIAN_INCOME'] = median_incomes.MEDIAN_INCOME.str.replace(',','')
median_incomes['MEDIAN_INCOME'] = median_incomes.MEDIAN_INCOME.str.replace('+','')
```

```python
median_incomes['MEDIAN_INCOME'] = median_incomes['MEDIAN_INCOME'].astype('int')
```

```python
median_incomes.loc[median_incomes['STATION']=='Myrtle, Halsey', 'STATION'] = 'Myrtle Av'
median_incomes.loc[median_incomes['STATION']=='Grand, Morgan', 'STATION'] = 'Grand St'
```

```python
median_incomes = median_incomes.append([median_incomes.loc[median_incomes['STATION']=='Myrtle Av'],
                      median_incomes.loc[median_incomes['STATION']=='Grand St']], ignore_index=True)
```

```python
#renaming the duplicates
median_incomes.loc[median_incomes.index == 22, 'STATION'] = 'Halsey St'
median_incomes.loc[median_incomes.index == 23, 'STATION'] = 'Morgan Av'
```

```python
#making station names uppercase
median_incomes['STATION'] = median_incomes['STATION'].str.upper()
```

```python
#shortening GEO_ID to sync with something else later
median_incomes['GEO_ID'] = median_incomes.GEO_ID.apply(lambda geo: geo[9:])

median_incomes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
L_stations = [
    '8th Av',
    '6th Av',
    'Union Square',
    '3rd Av',
    '1st Av',
    'Bedford Av',
    'Lorimer St', 
    'Graham Av', 
    'Grand St',
    'Montrose Av',
    'Morgan Av',
    'Jefferson St',
    'DeKalb Av',
    'Myrtle Av',
    'Halsey St',
    'Wilson Av',
    'Bushwick Av',
    'Broadway Jct',
    'Atlantic Av',
    'Sutter Av',
    'Livonia Av',
    'New Lots Av',
    'East 105th St',
    'Canarsie']

station_order = [x.upper() for x in L_stations]
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(20, 10))
plt.plot('STATION', 'MEDIAN_INCOME', data=median_incomes)
plt.xticks(rotation=30)
plt.xlabel('L Train Stations', fontsize='17', horizontalalignment='center')
plt.ylabel('Median Household Income (yearly)', fontsize='17', horizontalalignment='center')
plt.title('Median Household Income near L Train Stations (from 2019 census data)', fontsize='26')
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(20, 10))
income_graph = sns.barplot(x='STATION', y='MEDIAN_INCOME', data=median_incomes, order = station_order)
plt.xticks(rotation=30)
plt.xlabel('L Train Stations', fontsize='17', horizontalalignment='center')
plt.ylabel('Median Household Income (yearly)', fontsize='17', horizontalalignment='center')
plt.title('Median Household Income near L Train Stations (from 2019 census data)', fontsize='26')
```
